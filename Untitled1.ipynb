```python
nltk.download('stopwords')
from nltk.stem import WordNetLemmatizer
lemmatizer=WordNetLemmatizer()
stopwords=nltk.corpus.stopwords.words('english')
nltk.download('wordnet')
```

```python
#function for prepocessing our data
def preprocess_text(text):
  #1.remove punctuations 2.remove stop words 3.convert word into lowercase 4.tokenize the word 5.lemmatize
  text=text.lower()
  punc=''.join([word.lower() for word in text if word not in string.punctuation])
  tokens=re.split('\W+',punc)
  text=[lemmatizer.lemmatize(word) for word in tokens if word not in stopwords]
  return ' '.join(text)
```

```python
df['v2']=df['v2'].apply(lambda x: (preprocess_text(x)))
df.head()
```

```python
df['v1']=df['v1'].map({'ham':0,'spam':1})
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(analyzer='word')
df_1=cv.fit_transform(df['v2'])
df_dtm=pd.DataFrame(df_1.toarray(),columns=cv.get_feature_names())
df_dtm
```

```python
x=df_dtm
y=df['v1']
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
print(x_train,x_test)
```

```python
from sklearn.naive_bayes import MultinomialNB
clf=MultinomialNB()
clf.fit(x_train,y_train)
predict=clf.predict(x_test)
clf.score(x_train,y_train)
```

```python
from sklearn.metrics import plot_confusion_matrix
plot_confusion_matrix(clf,x_train,y_train,values_format='d')
```
